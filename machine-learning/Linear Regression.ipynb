# Linear Regression
- **Linear Regression** is a type of supervised learning algorithm that models the relationship between a dependent (target) variable and one or more independent (input) variables.
- The goal is to predict the value of the dependent variable based on the values of the independent variables.
- There are two main types of linear regression:
  - **Simple Linear Regression**: Involves one independent variable.
  - **Multiple Linear Regression**: Involves more than one independent variable.


## Simple Linear Regression
- **Simple Linear Regression** attempts to find the linear relationship between a dependent variable $y$ and a single independent variable $x$.
- The relationship is modeled with a straight line that best fits the data points.

### **Equation:**
$y = m \cdot x + b$

Where:
- $y$ = dependent variable (the output we are trying to predict).
- $x$ = independent variable (the input feature).
- $m$ = slope or coefficient (how much $y$ changes with a unit change in $x$).
- $b$ = intercept (the value of $y$ when $x = 0$).

### **Objective:**
- The goal is to determine the values of $m$ and $b$ such that the line fits the data points as closely as possible.

### **Example:**
- Consider predicting a person's salary ($y$) based on their years of experience ($x$). The relationship between the two can be modeled with the equation $y = m \cdot x + b$.

## Multiple Linear Regression
- **Multiple Linear Regression** is an extension of Simple Linear Regression. It models the relationship between one dependent variable $y$ and multiple independent variables $x_1, x_2, \dots, x_n$.
- This technique is used when we have more than one feature affecting the target variable.

### **Equation:**
$y = \beta_0 + \beta_1 \cdot x_1 + \beta_2 \cdot x_2 + \dots + \beta_n \cdot x_n
$

Where:
- $y$ = dependent variable (the outcome we want to predict).
- $x_1, x_2, \dots, x_n$ = independent variables (features that affect $y$).
- $\beta_0$ = intercept (the value of $y$ when all $x_i = 0$).
- $\beta_1, \beta_2, \dots, \beta_n$ = coefficients for each corresponding independent variable.

### **Objective:**
- The goal is to find the values of $\beta_0, \beta_1, \dots, \beta_n$ that best predict $y$ by minimizing the prediction error.

### **Example:**
- Suppose you want to predict house prices ($y$) based on the number of bedrooms ($x_1$), square footage ($x_2$), and age of the house ($x_3$). Multiple Linear Regression would help you find the relationship between these features and the house price.

## Cost Function: Mean Squared Error (MSE)
- In both Simple and Multiple Linear Regression, we use a **cost function** to measure how well the model predicts the target variable.
- One common cost function is the **Mean Squared Error (MSE)**, which calculates the average of the squared differences between the predicted values $\hat{y}$ and the actual values $y$.

### **MSE Formula:**
$MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$

Where:
- $n$ = the number of data points.
- $y_i$ = the actual value of the $i^{th}$ data point.
- $\hat{y}_i$ = the predicted value of the $i^{th}$ data point.

### **Interpretation:**
- The MSE gives a measure of the average squared difference between the actual and predicted values. A smaller MSE indicates that the model’s predictions are closer to the actual values.
- Our goal is to minimize the MSE so that the line fits the data as closely as possible.

## Gradient Descent Optimization
- To minimize the cost function (like MSE), we use an optimization algorithm called **Gradient Descent**.
- Gradient Descent iteratively updates the model’s parameters (such as $m$ and $b$) by moving in the direction that reduces the error (i.e., in the direction of the steepest slope).

### **Gradient Descent Formula:**
$
\beta_j = \beta_j - \alpha \cdot \frac{\partial}{\partial \beta_j} J(\beta)
$

Where:
- $\beta_j$ = the parameter being updated (either a coefficient or intercept).
- $\alpha$ = the learning rate, which controls how big the step is at each iteration.
- $J(\beta)$ = the cost function (e.g., MSE).
- $\frac{\partial}{\partial \beta_j}$ = the gradient or slope of the cost function with respect to $\beta_j$.

### **Simple Linear Regression Updates:**
For the intercept ($b$) and slope ($m$) in Simple Linear Regression, the updates look like this:

$
b = b - \alpha \cdot \frac{1}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i)
$
$
m = m - \alpha \cdot \frac{1}{n} \sum_{i=1}^{n} (\hat{y}_i - y_i) \cdot x_i
$

Where:
- $m$ = slope (coefficient) of the line.
- $b$ = intercept of the line.
- $n$ = number of training samples.

### **Objective:**
- The algorithm iteratively updates the parameters until the cost function reaches a minimum, which corresponds to the best-fitting line.

## Key Concepts

### **Overfitting**
- **Overfitting** occurs when the model fits the training data too well, capturing noise and fluctuations, but performs poorly on new, unseen data.
- This happens when the model is too complex (e.g., too many features or a very flexible model).

### **Underfitting**
- **Underfitting** occurs when the model is too simple and fails to capture the underlying pattern in the data.
- This happens when the model doesn't use enough features or has too rigid a structure.

### **Regularization**
- **Regularization** techniques, like **Lasso** (L1 regularization) and **Ridge** (L2 regularization), are used to prevent overfitting by adding a penalty term to the cost function.
- This penalty discourages large coefficient values, helping the model generalize better to unseen data.

## Conclusion
- **Linear Regression** is one of the simplest and most interpretable machine learning models.
- It is effective for predicting continuous values and is easy to understand because of its linear nature.
- By carefully tuning the model and avoiding overfitting or underfitting, Linear Regression can be a powerful tool for predicting outcomes based on data.


# Implementation of Simple Linear Regression

We have a house prices data set contains of area in square feet and price in USD and we are going to predict the house price by passing area to linear regression model

```python
import pandas as pd
```

Using pandas to read csv files

```python
df = pd.read_csv('data/home_prices.csv')
df
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Area (Sqr Ft)')
plt.ylabel('Price (USD)')
plt.scatter(df.area,df.price)
```

Ploting points on graph to projecting how relation are

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
X = df[['area']]
y = df['price']
reg.fit(X,y)
```

```python
reg.predict([[3300]])
```
