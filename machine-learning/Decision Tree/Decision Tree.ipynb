```python
import pandas as pd
df = pd.read_csv('data/salaries.csv')
df.head()
```

```python
inputs = df.drop('salary_more_then_100k',axis='columns')
target = df['salary_more_then_100k']
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_company.fit_transform(inputs['job'])
inputs['degree_n'] = le_company.fit_transform(inputs['degree'])
inputs
```

```python
inputs_n = inputs.drop(['company','job','degree'],axis='columns')
inputs_n.head()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(inputs_n,target)
```

```python
model.score(inputs_n,target)
```

**Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?**

```python
model.predict([[2,1,0]])
```

**Is salary of Google, Computer Engineer, Masters degree > 100 k ?**

```python
model.predict([[2,1,1]])
```

# Exercise: Build decision tree model to predict survival based on certain parameters
![](https://raw.githubusercontent.com/codebasics/py/801ee0ee4d342fd22b89915dc0c4864250a0ec10/ML/9_decision_tree/titanic.jpg)

CSV file is available to download at [DATA](https://github.com/codebasics/py/blob/master/ML/9_decision_tree/Exercise/titanic.csv)

In this file using following columns build a model to predict if person would survive or not,
1. Pclass
2. Sex
3. Age
4. Fare

Calculate score of your model

```python
titanic_data = pd.read_csv("data/titanic.csv")
titanic_data.head()
```

```python
titanic_data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
titanic_data.head()
```

```python
inputs = titanic_data.drop('Survived',axis='columns')
target = titanic_data['Survived']
```

```python
inputs['Sex'] = inputs.Sex.map({'male': 1, 'female': 2})
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
```

```python
inputs.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict([[3,2,2,5]])
```
